```python
# class tasks

cgpa = float(input("Enter your CGPA: "))

if 4.0 >= cgpa >= 3.8:
    print(f"Your grade is A+ and your CGPA is {cgpa}")
elif 3.8 > cgpa >= 3.6:
    print(f"Your grade is A and your CGPA is {cgpa}")
elif 3.6 > cgpa >= 3.2:
    print(f"Your grade is B+ and your CGPA is {cgpa}")
elif 3.2 > cgpa >= 2.8:
    print(f"Your grade is B and your CGPA is {cgpa}")
elif 2.8 > cgpa >= 2.4:
    print(f"Your grade is C+ and your CGPA is {cgpa}")
elif 2.4 > cgpa >= 2.0:
    print(f"Your grade is C and your CGPA is {cgpa}")
elif 2.0 > cgpa >= 1.6:
    print(f"Your grade is D+ and your CGPA is {cgpa}")
elif 1.6 > cgpa >= 1.0:
    print(f"Your grade is D and your CGPA is {cgpa}")
elif 1.0 > cgpa >= 0:
    print(f"Your grade is F and your CGPA is {cgpa}")
else:
    print("Invalid CGPA (NG)")
```
