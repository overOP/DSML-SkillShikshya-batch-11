```python
x = int(input("Enter a number: "))
z = int(input("Enter a number: "))
y = input("Enter a sumb: ")
def udf(x,z,y):
   if y == "-":
    return f"this is the total: {x - z}"
   elif y == "+":
    return f"this is the total: {x + z}"
   elif y == "*":
    return f"this is the total: {x * z}"
   elif y == "/":
    return f"this is the total: {x / z}"
   else:
    return "the is not '-,+,*,/"
udf(x,z,y)
```
