```python
po_lst = []
ne_lst = []

for i in range(5):
    num = int(input("Enter a number: "))

    if num > 0:
        po_lst.append(num)
    elif num < 0:
        ne_lst.append(num)
    else:
        break

print("Positive numbers:", po_lst)
print("Negative numbers:", ne_lst)
```
