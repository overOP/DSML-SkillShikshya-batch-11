```python
#deep copy and shallow copy
lst = [1,3,5,6]
lst1 = lst
print(lst1)
lst2 = lst1.copy()
lst2.append(7)
print(lst2)
print(lst1)
```

```python
import copy

a = [[1,2], [3,4]]

# shallow copy
b = a.copy()

# deep copy
c = copy.deepcopy(a)

a[0].append(99)

print(a)
print(b)
print(c)
```
